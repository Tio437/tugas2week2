```python

# k singkatan kalkulus
# ms sungkatan matimatika statitiska

shafira_k=85
amanda_k=80
aditya_k=75
nedia_k=95
widya_k=85
hanif_k=75
andi_k=70
dhanar_k=85
hikma_k=80
shafira_ms=75
amanda_ms=90
aditya_ms=80
nedia_ms=80
widya_ms=85
hanif_ms=95
andi_ms=75
dhanar_ms=85
hikma_ms=75

#rr singkatan rata rata
#n singkatan nilai

# rata rata nilai shafira
rr_n_shafira=(shafira_k+shafira_ms) /2
# jumplah nilai hanif dan andi
jmplh_n_hanif_dan_andi=hanif_k+hanif_ms+andi_k+andi_ms
# rata rata nilai kalkulus widya dhanar hikma dan nedia
rr_n_kalkulus_widya_dhanar_hikma_dan_nedia=(widya_k+dhanar_k+hikma_k+nedia_k) /4
# rata rata  nilai metode statitiska widya dhanar hikma dan nedia
rr_n_statitiska_widya_dhanar_hikma_dan_nedia=(widya_ms+dhanar_ms+hikma_ms+nedia_ms) /4
# rata rata nilai mata kuliah kalkulus
rr_n_mata_kuliah_kalkulus=(shafira_k+amanda_k+aditya_k+nedia_k+widya_k+hanif_k+andi_k+dhanar_k+hikma_k) /9
# rata rata nilai mata kuliah metode statitiska
rr_n_mata_kuliah_statitiska=(shafira_ms+amanda_ms+aditya_ms+nedia_ms+widya_ms+hanif_ms+andi_ms+andi_ms+dhanar_ms+hikma_ms) /9


print("rata rata nilai shafira=",rr_n_shafira)
print("jumplah nilai hanif dan andi=",jmplh_n_hanif_dan_andi)
print("rata rata nilai kalkulus widya dhanar hikma dan nedia=",rr_n_kalkulus_widya_dhanar_hikma_dan_nedia)
print("rata rata nilai statitiska widya dhanar hikma dan nedia=", rr_n_statitiska_widya_dhanar_hikma_dan_nedia)
print("rata rata nilai mata kuliah Kalkulus=", rr_n_mata_kuliah_kalkulus)
print("rata rata nilai mata kuliah Metode statitiska=", rr_n_mata_kuliah_statitiska)
```

```python
lama_kerja=10 #tahun
jamkerja=8 #jam
gaji_perjam=30000
satu_bulan=30 #hari
tunjangan_perlimatahun=10/100
tunjangan_ini=5 #tahun< akan mendapatkan tunjangan
tidak_masuk_kerja=3 #hari pada bulan agustus
upah_lembur=10000 #perjam
lembur=5 #jam


#total gaji 1 hari
gaji_sehari=jamkerja*gaji_perjam

#gaji bulan agustus
gaji_bulan=(satu_bulan-tidak_masuk_kerja)*gaji_sehari
print('Gaji Dwi pada bulan Agustus=Rp',gaji_bulan)
#tunjangan bulan agustus
if tunjangan_ini<=lama_kerja :
  tunjangan=tunjangan_perlimatahun*gaji_bulan
else:
  tunjangan=0
print('Total tunjangan Dwi pada bulan Agustus=Rp',tunjangan)
#Total gaji pada bulan Agustus di tambah lembur
total_gaji_dan_lembur=gaji_bulan+(lembur*upah_lembur)
print('Gaji Dwi bulan Agustus + lembur=Rp',total_gaji_dan_lembur)
#Total gaji pada bulan Agustus di tambah lembur dan setelah membayar tunjangan
total_gaji_lembur_tunjangan=total_gaji_dan_lembur+tunjangan
print('Gaji Dwi bulan Agustus + lembur dan di tambah tunjangan=Rp',total_gaji_lembur_tunjangan)



```
